```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

# Задана функція y = 2 + ln(4 + sin(x))
def задана_функція(x):
    return 2 + np.log(4 + np.sin(x))

# Функція для глобальної інтерполяції
def глобальна_інтерполяція(x, y, x_new):
    поліном = np.poly1d(np.polyfit(x, y, len(x) - 1))
    return поліном(x_new)

# Функція для локальної лінійної інтерполяції
def локальна_лінійна_інтерполяція(x, y, x_new):
    f = interp1d(x, y, kind='linear')
    return f(x_new)

# Розрахунок відносної похибки
def відносна_похибка(y_апрокс, y_фактичні):
    return np.abs((y_апрокс - y_фактичні) / y_фактичні) * 100

# Визначення інтервалу
інтервал = np.linspace(0, 2 * np.pi, 100)
y_фактичні = задана_функція(інтервал)

# Вибір вузлів для глобальної інтерполяції: кінцеві точки та ще три точки
вузли_x = np.array([інтервал[0], np.pi/2, np.pi, 3*np.pi/2, інтервал[-1]])  # приклад вузлів
вузли_y = задана_функція(вузли_x)

# Інтерполяція та розрахунок відносних похибок для довільної точки, не співпадаючої з вузлами
x_тест = np.pi/4  # Довільна точка для тестування
y_тест_фактичні = задана_функція(x_тест)

y_глобально_інтерпольовані = глобальна_інтерполяція(вузли_x, вузли_y, x_тест)
y_локально_інтерпольовані = локальна_лінійна_інтерполяція(вузли_x, вузли_y, x_тест)

похибка_глобальна = відносна_похибка(y_глобально_інтерпольовані, y_тест_фактичні)
похибка_локальна = відносна_похибка(y_локально_інтерпольовані, y_тест_фактичні)

# Побудова графіків
plt.figure(figsize=(14, 7))

# Графік фактичної функції
plt.plot(інтервал, y_фактичні, label='Фактична Функція')

# Графік глобальної інтерполяції
plt.plot(інтервал, глобальна_інтерполяція(вузли_x, вузли_y, інтервал), label='Глобальна Інтерполяція')

# Графік локальної лінійної інтерполяції
plt.plot(інтервал, локальна_лінійна_інтерполяція(вузли_x, вузли_y, інтервал), label='Локальна Лінійна Інтерполяція')

# Виділення вузлів інтерполяції
plt.scatter(вузли_x, вузли_y, color='red', zorder=5)

# Виділення тестової точки
plt.scatter(x_тест, y_тест_фактичні, color='green', zorder=5, label=f'Тестова Точка (x={x_тест:.2f}, y={y_тест_фактичні:.2f})')

# Відображення похибок на графіку
plt.text(1, 0, f'Похибка Глобальної Інтерполяції при x={x_тест}: {похибка_глобальна:.2f}%', fontsize=9, color='blue')
plt.text(1, -0.5, f'Похибка Локальної Інтерполяції при x={x_тест}: {похибка_локальна:.2f}%', fontsize=9, color='blue')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Інтерполяція Функції та Розрахунок Похибок')
plt.grid(True)
plt.show()

(похибка_глобальна, похибка_локальна)

```
